### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data.head()
```

```python
unique_data = purchase_data.nunique()
unique_data #prints number of unique values in each column
```

```python
total_players = unique_data["SN"]
#total_players = pd.DataFrame([total_players], columns = ["Total Players"])
print("Total number of players is:")
print(total_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data['Item ID'].nunique()
#unique_items prints the number of unique items in the Item ID column
avg_price = purchase_data["Price"].mean()
#avg_price prints the average price
num_purch = len(purchase_data["Item Name"])
#num_purch prints total number of purchases
total_rev = avg_price * num_purch
#total_rev prints total revenue

sum_data = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [avg_price], "Number of Purchases": [num_purch],
             "Total Revenue": [total_rev]})
sum_data
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_group = purchase_data[["SN", "Gender"]]
gender_group = gender_group.drop_duplicates()
#gender_group #groups the SN and Gender columns

gender_counts = gender_group["Gender"].value_counts()
#gender_counts #prints number of each gender

total_count = [gender_counts[0], gender_counts[1], gender_counts[2]]
percent_gender = [round((gender_counts[0]/total_players)*100,2),
                  round((gender_counts[1]/total_players)*100,2),
                round((gender_counts[2]/total_players)*100,2)] #calculates the gender percent
#percent_gender #prints percent of each gender

gender_dems = pd.DataFrame({"Percentage of Players": percent_gender, "Total Count": gender_counts})
gender_dems.index = (["Male", "Female", "Other / Non-Disclosed"])
gender_dems #prints count and percent of genders
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
group_gen = purchase_data[["SN","Gender","Price"]]
group_gen = group_gen.drop_duplicates()
count_gender = group_gen["Gender"].value_counts()

purch_count = [count_gender[0],count_gender[1],count_gender[2]]
purch_count #prints purchase count

group_gen = group_gen.groupby("Gender")
total_spent = group_gen.sum()
total_spent #prints total purchase value

total_value = [total_spent.iloc[1,0], total_spent.iloc[0,0], total_spent.iloc[2,0]]

avg_purch = [total_spent.iloc[1,0]/count_gender[0], total_spent.iloc[0,0]/count_gender[1],
             total_spent.iloc[2,0]/count_gender[2]]
avg_purch #prints average purchase price

norm_total = [total_spent.iloc[1,0]/purch_count[0], total_spent.iloc[0,0]/purch_count[1],
              total_spent.iloc[2,0]/purch_count[2]]
#norm_total #avg total purchase per person????

purch_analysis = pd.DataFrame({"Purchase Count": purch_count, "Average Purchase Price": avg_purch,
                                  "Total Purchase Value": total_value, "Avg Total Purchase per Person": norm_total, "Gender": ["Male", "Female", "Other"]})
purch_analysis = purch_analysis.set_index("Gender")
purch_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_df = pd.DataFrame(purchase_data)

bins = [0, 9, 14, 19, 24, 29, 34, 39, 100] #creates bins
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] #creates bin labels
age_group = age_df.groupby(pd.cut(age_df["Age"], bins, labels=bin_labels)) #slices the data at the Age column
#age_group #prints object of bins

age_dem = pd.DataFrame({"Purchase Count": age_group["Price"].count(), "Average Purchase Price": age_group["Price"].mean(),
                           "Total Purchase Value": age_group["Price"].sum(), "Avg Total Purchase per Person": age_group["Price"].sum()/age_group["SN"].nunique()})
age_dem
```
